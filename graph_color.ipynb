A graph G with 15 vertices, which are numbered from 1 to 15, is given. The vertices with numbers i and j are connected by an edge if |i−j|≤3, and are not connected otherwise. We call a coloring of graph vertices in black and white "good" if exactly 6 vertices are colored in white, and the number of edges whose ends are colored in different colors is even. How many good colorings of the graph G are there?

```python
import numpy
```

```python
c=0
G=[] #constructing a graph
for i in range(1,16):
  for j in range(i+1,16):
      if (j-i)<4:
        #c=c+1 #number of edges
        G.append((i,j))
#print(c)
#print(G)


```

```python
from itertools import combinations
l = numpy.array(range(1,16)) #all subsets of 6 = all possible colorings of a graph
print([j for j in combinations(l,6)])
subsets=[j for j in combinations(l,6)]
print(subsets)
print(len(subsets))
```

```python
f=0 #choosing a graph with a right coloring
for color in subsets:
  k=0
  for g in G:
    if (len((set(color)).intersection(set(g)))==1):
      k=k+1
  if (k % 2)==0:
    f=f+1

print(f)

```
